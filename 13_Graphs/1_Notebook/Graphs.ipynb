#### Multiple graphs on the same set of Axes

```python
x2 = np.linspace(0,10,100)
y2 = np.sin(x2)
```

```python
plt.figure(figsize=(10,5))

plt.plot(x,y,'o', color='purple', lw='0.7', ms='5', label='component 1')
plt.plot(x2,y2,'-', color='green', lw='1', ms='5', label='component 2')

plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')

plt.legend(loc='lower right', fontsize="10")

plt.show()
```

### Manipulating the axis

```python
plt.figure(figsize=(10,5))

plt.plot(x,y,'o', color='purple', lw='0.7', ms='5', label='component 1')
plt.plot(x2,y2,'-', color='green', lw='1', ms='5', label='component 2')

plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')

plt.legend(loc='lower right', fontsize="10")

plt.xticks([2,4,6,8])
plt.yticks([-2,-1,0,1,2])

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(x,y,'o', color='purple', lw='0.7', ms='5', label='Data')
plt.plot(x2,y2,'-', color='green', lw='1', ms='5', label='True')

plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')

plt.legend(loc='upper right', ncol=2, fontsize="10")

plt.xticks([2,4,6,8])
plt.yticks([-2,-1,0,1,2,6])

plt.show()
```

### Histogram

```python
res = np.random.randn(1000)*0.2 + 0.4
res
```

```python
a = plt.hist(res)

plt.show()

a
```

a returns two objects

The first object is the number of values in the first interval (bin)

The second array is a list of the endpoints for the intervals (bins)

```python
plt.hist(res, bins=100)

plt.show()
```

```python
plt.figure(figsize=(8,3))

plt.hist(res, bins=30)

plt.show()
```

### Normalized Histogram

The area under of all the bars (bins) add up to 1

```python
plt.figure(figsize=(8,3))

plt.hist(res, bins=30, density=True)

plt.show()
```

### Different Style of Histograms

```python
res2 = np.random.randn(1000)*0.2 + 0.4

plt.figure(figsize=(8,3))

plt.hist(res, bins=30, density=True)
plt.hist(res2, bins=30, density=True)

plt.show()
```

Cannot see the blue histogram which is covered by the green histogram. We require a better way to visualize this.

```python
res2 = np.random.randn(1000)*0.2 + 0.4

plt.figure(figsize=(8,3))

plt.hist(res, bins=30, density=True, histtype='step')
plt.hist(res2, bins=30, density=True, histtype='step')

plt.show()
```

#### Adding Text to your Graph

```python
figure, axes = plt.subplots(2,2, figsize=(15,10))

ax00 = axes[0][0]
ax00.plot(x,y, label='fake-sin-graph')
ax00.set_xlabel('fake-x-data')
ax00.set_ylabel('fake-sin-graph')
ax00.set_title('Graph00')
ax00.legend()

ax01 = axes[0][1]
ax01.plot(x2,y2, label='real-sin-graph')
ax01.set_xlabel('x-data')
ax01.set_ylabel('ideal-sin-graph')
ax01.set_title('Graph01')
ax01.legend(loc='lower left')

ax10 = axes[1][0]
ax10.hist(res)
ax10.text(0.1,0.1, 'This is a string of text', fontsize=20, color='yellow')

ax11 = axes[1][1]
ax11.plot(res)

plt.show()
```

#### Change Axis fontsize and Labels

```python
figure, axes = plt.subplots(2,2, figsize=(15,10))

ax00 = axes[0][0]
ax00.plot(x,y, label='fake-sin-graph')
ax00.set_xlabel('fake-x-data')
ax00.set_ylabel('fake-sin-graph')
ax00.set_title('Graph00')
ax00.legend()
ax00.tick_params(axis='both', labelsize=30)

ax01 = axes[0][1]
ax01.plot(x2,y2, label='real-sin-graph')
ax01.set_xlabel('x-data')
ax01.set_ylabel('ideal-sin-graph')
ax01.set_title('Graph01')
ax01.legend(loc='lower left')

ax10 = axes[1][0]
ax10.hist(res)
ax10.text(0.1,0.1, 'This is a string of text', fontsize=20, color='yellow')

ax11 = axes[1][1]
ax11.plot(res)

plt.show()
```
