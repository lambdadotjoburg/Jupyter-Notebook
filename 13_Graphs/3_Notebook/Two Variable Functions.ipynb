## Functions of Two Variables

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

```python
interval = np.linspace(-1,1,100)
x,y = np.meshgrid(interval,interval)
z = x**2+x*y
```

#### Contour Plots

```python
plt.contour(x,y,z, levels=30)
plt.show()
```

#### Filled-In Contour Plots

```python
plt.contourf(x,y,z, levels=30)
plt.show()
```

```python
plt.contourf(x,y,z, levels=30)
plt.colorbar(label='color bar')
plt.rcParams['axes.grid'] = False
plt.show()
```

```python
plt.contourf(x,y,z, levels=30, vmin=0.7)
plt.colorbar(label='color bar')
plt.rcParams['axes.grid'] = False # This line removes an error for some deprecated package stuff
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.contourf(x,y,z, levels=30)
plt.colorbar(label='color bar')

plt.show()
```

```python
plt.contourf(x,y,z, levels=30, vmax=1.20, vmin=0.48)

plt.colorbar(label='color bar')
plt.rcParams['axes.grid'] = False # This line removes an error for some deprecated package stuff

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.contourf(x,y,z, levels=30, cmap='plasma')

plt.colorbar(label='color bar')
plt.rcParams['axes.grid'] = False # This line removes an error for some deprecated package stuff

plt.show()
```

#### Sub Contour Plots

```python
figure, axes = plt.subplots(2,2, figsize=(15,10))

ax00 = axes[0][0]
ax00.contourf(x,y,z, levels=30)

ax01 = axes[0][1]
ax01.contourf(x,y,z, levels=30, cmap='plasma')

ax10 = axes[1][0]
ax10.contour(x,y,z, levels=30)

ax11 = axes[1][1]
ax11.contour(x,y,z, levels=30, cmap='plasma')

plt.show()
```

```python
plt.figure(figsize=(10,7))

plt.rcParams['axes.grid'] = True

cplt = plt.contour(x,y,z, levels=10)

plt.clabel(cplt, fontsize=10)

plt.show()
```
