```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

## Gradient/Vector Field

```python
_ = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(_,_)

U = -1 - X**2 + Y
V = 1 + X - Y**2

# Speed is the magnitude of a vector

speed = np.sqrt(U**2 + V**2)
```

```python
figure, axes = plt.subplots(2, 2, figsize=(10,10))

ax00 = axes[0][0]
ax00.streamplot(X,Y,U,V)

ax01 = axes[0][1]
ax01.streamplot(X,Y,U,V, color=speed)

lw=5*speed/speed.max()
ax10 = axes[1][0]
ax10.streamplot(X,Y,U,V, linewidth=lw)

ax11 = axes[1][1]
ax11.streamplot(X,Y,U,V)

plt.show()
```

```python
_ = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(_,_)

U = -1 - X**2 + Y
V = 1 + X - Y**2

# Speed is the magnitude of a vector

speed = np.sqrt(U**2 + V**2)
```

```python
figure, axes = plt.subplots(2, 2, figsize=(10,10))

ax00 = axes[0][0]
ax00.streamplot(X,Y,U,V)

ax01 = axes[0][1]
ax01.streamplot(X,Y,U,V, color=speed)

lw=5*speed/speed.max()
ax10 = axes[1][0]
ax10.streamplot(X,Y,U,V, linewidth=lw)

ax11 = axes[1][1]
ax11.quiver(X,Y,U,V, color="C0", angles='xy', scale_units='xy', scale=15, width=.0055)

plt.show()
```
