# Graphs

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science', 'notebook', 'grid'])
```

```python
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

```python
plt.hist(res_a1, bins=50)
plt.hist(res_b1, bins=50)
plt.hist(res_a2, bins=50)
plt.hist(res_b2, bins=50)

plt.show()
```

```python
mySeparator = '\n' # this goes to a new line

textstr1 = mySeparator.join((
    r'$\sigma_a=$%.4f' % (np.std(res_a1)),
    r'$\sigma_b=$%.4f' % (np.std(res_b1))
))

textstr2 = mySeparator.join((
    r'$\sigma_a=$%.4f' % (np.std(res_a2)),
    r'$\sigma_b=$%.4f' % (np.std(res_b2))
))
```

```python
textstr1
```

```python
figures, axes = plt.subplots(1,2, figsize=(15,6))

ax0 = axes[0]
ax0.hist(res_a1, bins=30, density=True, histtype='step', label='a')
ax0.hist(res_b1, bins=30, density=True, histtype='step', label='b')
ax0.legend(loc='upper right')
ax0.text(-0.25,1.8, textstr1, fontsize=15)

plt.show()
```

```python
figures, axes = plt.subplots(1,2, figsize=(15,6))

ax0 = axes[0]
ax0.hist(res_a1, bins=30, density=True, histtype='step', label='a')
ax0.hist(res_b1, bins=30, density=True, histtype='step', label='b')
ax0.legend(loc='upper right')
ax0.text(0.05,0.81, textstr1, fontsize=15, transform=ax0.transAxes)

plt.show()
```

```python
figures, axes = plt.subplots(1,2, figsize=(15,6))

ax0 = axes[0]
ax0.hist(res_a1, bins=30, density=True, histtype='step', label='a')
ax0.hist(res_b1, bins=30, density=True, histtype='step', label='b')
ax0.legend(loc='upper right')
ax0.text(0.05,0.81, textstr1, fontsize=15, transform=ax0.transAxes, bbox=dict(facecolor='white', edgecolor='black'))

plt.show()
```

```python
figures, axes = plt.subplots(1,2, figsize=(15,6))

ax0 = axes[0]
ax0.hist(res_a1, bins=30, density=True, histtype='step', label='a')
ax0.hist(res_b1, bins=30, density=True, histtype='step', label='b')
ax0.legend(loc='upper right', fancybox=False, edgecolor='black')
ax0.text(0.05,0.81, textstr1, fontsize=15, transform=ax0.transAxes, bbox=dict(facecolor='white', edgecolor='black'))

plt.show()
```

```python
figures, axes = plt.subplots(1,2, figsize=(15,6))

ax0 = axes[0]
ax0.hist(res_a1, bins=30, density=True, histtype='step', label='a')
ax0.hist(res_b1, bins=30, density=True, histtype='step', label='b')
ax0.legend(loc='upper right', fancybox=False, edgecolor='black')
ax0.text(0.05,0.81, textstr1, fontsize=15, transform=ax0.transAxes, bbox=dict(facecolor='white', edgecolor='black'))
ax0.set_ylabel('Density')
figures.text(0.5, -0.02, '$\Delta E \quad [Joules]$', ha='center', fontsize=20)

plt.show()
```

```python
figures, axes = plt.subplots(1,2, figsize=(15,6))

ax0 = axes[0]
ax0.hist(res_a1, bins=30, density=True, histtype='step', label='a', color='blue')
ax0.hist(res_b1, bins=30, density=True, histtype='step', label='b', color='red')
ax0.legend(loc='upper right', fancybox=False, edgecolor='black')
ax0.text(0.05,0.81, textstr1, fontsize=15, transform=ax0.transAxes, bbox=dict(facecolor='white', edgecolor='black'))
ax0.set_ylabel('Density')
ax0.set_title('Trial 1')

ax1 = axes[1]
ax1.hist(res_a1, bins=30, density=True, histtype='step', label='a', color='blue')
ax1.hist(res_b1, bins=30, density=True, histtype='step', label='b', color='red')
ax1.legend(loc='upper right', fancybox=False, edgecolor='black')
ax1.text(0.05,0.81, textstr2, fontsize=15, transform=ax1.transAxes, bbox=dict(facecolor='white', edgecolor='black'))
ax1.set_title('Trial 2')

figures.text(0.5, -0.02, '$\Delta E \quad [Joules]$', ha='center', fontsize=20)

figures.text(0.52, 0.95, '$Experiment$', ha='center', fontsize=20)

plt.show()
```
