# Graphing with <code style='background: white'>matplotlib</code>

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parabola

Turning Point form of Parabola

$$ y = (x-p)^2 + q$$

```python
a=1
p=2
q=-1

Dx = np.linspace(-1,5,100) # Specify independent variable as a domain with interval step size

# Define relationship between independent and dependent variable

def f(x):
    return a*(x-p)**2 + q

fig = plt.figure(figsize=(6,5))

ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')


plt.xticks([-1,0,1,2,3,4,5])
plt.yticks([-1,0,1,2,3,4,5,6,7,8])


plt.plot(Dx, f(Dx)) # The graph is restricted to the specified domain


plt.show()
```

Standard Form of Parabola

$$f(x) = a x^2 + b x + c$$

```python
a=1
b=2
c=1

Dx = np.linspace(-3,2,100) # Specify independent variable as a domain with interval step size

# Define relationship between independent and dependent variable

def f(x):
    return a*x**2 + b*x**1 + c*x**0

fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xticks([-1,0,1,2,3,4,5])
plt.yticks([-1,0,1,2,3,4,5])

plt.plot(Dx, f(Dx)) # The graph is restricted to the specified domain

plt.show()
```

## Hyperbola

$$ y = \frac{k}{x-p}+q $$

```python
# Parameters

p = 2
q = 3
k = 1

# Vertical and Horizontal Asymptotes

VA = p
HA = q

x1 = np.linspace(-5+VA,-1/5+VA,100) # Translate by VA units
x2 = np.linspace(1/5+VA,5+VA,100) # Translate by VA units

y1 = k/(x1-VA) + HA # Define relationship between independent and dependent variable
y2 = k/(x2-VA) + HA # Define relationship between independent and dependent variable

l = np.linspace(-5+VA,5+VA,2) # Translate by VA units
o = np.linspace(0+HA,0+HA,2) # Translate by HA units

r = np.linspace(0+VA,0+VA,2) # Translate by VA units
s = np.linspace(-5+HA,5+HA,2) # Translate by HA units



fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.rcParams['figure.figsize'] = 2,2
plt.rcParams['axes.grid'] = True



plt.plot(x1, y1,'b-') # The graph is restricted to the specified domain
plt.plot(x2, y2,'b-') # The graph is restricted to the specified domain

plt.plot(l,o,'k--') # Horizontal Asymptote

plt.plot(r,s,'k--') # Vertical Asymptote

plt.show()
```
