```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

filename = "example_data.csv"

with open(filename, 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
```

```python
y
```

```python
plt.plot(x,y, label='loaded from file')

plt.xlabel('x')
plt.ylabel('y')
plt.title('interesting Graph')

plt.show()
```

```python
plt.plot(x,y,'o--')

plt.xlabel('x')
plt.ylabel('y')
plt.title('interesting Graph')

plt.show()
```
