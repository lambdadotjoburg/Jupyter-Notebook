# Trigonometric Graphs

```python
# Import appropriate python packages

from numpy import *
import matplotlib.pyplot as plt
import sympy as smp

plt.rcParams['axes.grid'] = True
```

### Define the Union Operation of Two Sets while Maintaining the Order

```python
def Union(lst1, lst2):
    final_list = list(set(lst1) | set(lst2))
    return sort(final_list)
```

### Define the Domain of the Function

```python
k=2

x = linspace(-2*pi, 2*pi, 1000)
```

### Generate a List for the Ticks along the x-Axis:

#### Step Size of $\pi / 3$ over the interval $(- 2 \pi, 2 \pi )$
#### Step Size of $\pi / 4$ over the interval $(- 2 \pi, 2 \pi )$
#### Take the Union of the above lists

```python
xvalsrads1 = list(arange(-2*smp.pi ,2*smp.pi+(smp.pi/6), (smp.pi/6)))
xvalsrads2 = list(arange(-2*smp.pi ,2*smp.pi, (smp.pi/4)))

xvalsrads = Union(xvalsrads1,xvalsrads2)
```

### Convert Radian Values in the List above to a List of Corresponding Degrees:

#### Step Size of $30^o$ over the interval $(-360^o, 360^o)$
#### Step Size of $45^o$ over the interval $(-360^o, 360^o)$
#### Take the Union of the above lists

```python
xvalsdegs1 = list(arange(-2*smp.pi*(360/(2*smp.pi)), ((2*smp.pi)+(smp.pi/6))*(360/(2*smp.pi)), (smp.pi/6)*(360/(2*smp.pi)) ))
xvalsdegs2 = list(arange(-2*smp.pi*(360/(2*smp.pi)), 2*smp.pi*(360/(2*smp.pi)), (smp.pi/4)*(360/(2*smp.pi)) ))

xvalsdegs = Union(xvalsdegs1,xvalsdegs2)
```

### Generate a List for the Ticks along the y-Axis:

#### Step Size of $\frac{\sqrt{3}}{2} $ over the interval $(0, k)$ and $\frac{\sqrt{3}}{2} $ over the interval $(0, -k)$
#### Step Size of $\frac{\sqrt{2}}{2} $ over the interval $(0, k)$ and $\frac{\sqrt{2}}{2} $ over the interval $(0, -k)$
#### Step Size of $\frac{1}{2} $ over the interval $(0, k)$ and $\frac{1}{2} $ over the interval $(0, -k)$
#### Take the Union of the above lists

```python
def Convert(lst):
    return [ -i for i in lst ]
```

```python
yvalspos01 = list(arange(0 , k, smp.sqrt(3)/2))
yvalsneg01 = Convert(yvalspos01)
yvals01 = list(Union(yvalsneg01,yvalspos01))
```

```python
yvalspos02 = list(arange(0 , k, smp.sqrt(2)/2))
yvalsneg02 = Convert(yvalspos02)
yvals02 = list(Union(yvalsneg02,yvalspos02))
```

```python
yvalspos03 = list(arange(0 , k, 1/2))
yvalsneg03 = Convert(yvalspos03)
yvals03 = list(Union(yvalsneg03,yvalspos03))
```

```python
endpoints = [-k, k]
```

```python
roots = list(Union(yvals01,yvals02))
halfs = list(Union(yvals03,endpoints))

yvals = list(Union(halfs,roots))
```

### Define the Trigonometric Functions

```python
def f(x):
    return sin(x)

def g(x):
    return cos(x)

def h(x):
    return tan(x)
```

```python
fig = plt.figure(figsize=(20,8))

ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xticks(xvalsrads,xvalsdegs, fontsize=11)
ax.set_yticks(yvals, yvals)

plt.xlabel("x", loc='right', fontsize=20)
plt.ylabel("y", loc='top', rotation=0, fontsize=20)

plt.rcParams['axes.grid'] = True

plt.plot(x, f(x), label=r'$f(x) = \sin(x)$')

plt.ylim(-k,k)

plt.legend(fontsize=20, loc='upper right')

plt.show()
```
