# Analytic Form of Circles

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Circle

$$ x^2 + y^2 = r^2 $$

```python
r = 1

x1 = np.linspace(0,r,100)
x2 = np.linspace(-r,0,100)

def f(x):
    return np.sqrt(r**2 - x**2)

plt.figure(figsize=(5,5))

plt.rcParams['axes.grid'] = True

plt.plot(x1, f(x1))
plt.plot(x1, -f(x1))

plt.plot(x2, f(x2))
plt.plot(x2, -f(x2))

plt.show()
```

## Ellipse at Center

$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 $$

```python
a = 1
b = 1

x1 = np.linspace(0,a,200)
x2 = np.linspace(-a,0,200)

def f(x):
    return b/a * np.sqrt(a**2-x**2)

plt.figure(figsize=(8,8))

plt.rcParams['axes.grid'] = True

plt.plot(x1, f(x1), 'b')
plt.plot(x1, -f(x1), 'g')

plt.plot(x2, f(x2), 'r')
plt.plot(x2, -f(x2), 'y')

plt.show()
```

## Ellipse at Center with Horizontal Major Axis

$$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 $$

```python
a = 2
b = 1

x1 = np.linspace(0,a,200)
x2 = np.linspace(-a,0,200)

def f(x):
    return b/a * np.sqrt(a**2-x**2)

plt.figure(figsize=(3*a,3*b))

plt.rcParams['axes.grid'] = True

plt.plot(x1, f(x1), 'b')
plt.plot(x1, -f(x1), 'g')

plt.plot(x2, f(x2), 'r')
plt.plot(x2, -f(x2), 'y')

plt.show()
```

## Hyperbola at Center with Horizontal Major Axis

$$ \frac{x^2}{a^2} - \frac{y^2}{b^2} = 1 $$

```python
a = 1
b = 1

x1 = np.linspace(a,5*a,200)
x2 = np.linspace(-5*a,-a,200)

def f(x):
    return b/a * np.sqrt(x**2 - a**2)

plt.figure(figsize=(3*a,3*b))

plt.rcParams['axes.grid'] = True

plt.plot(x1, f(x1), 'b')
plt.plot(x1, -f(x1), 'g')

plt.plot(x2, f(x2), 'r')
plt.plot(x2, -f(x2), 'y')

plt.show()
```
