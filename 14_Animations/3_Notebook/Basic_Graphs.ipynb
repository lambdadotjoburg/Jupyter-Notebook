## Hyperbola

```python
x = np.linspace(-5,5,100) # Specify independent variable as a domain with interval step size

y = 1/x # Define relationship between independent and dependent variable

plt.plot(x, y) # The graph is restricted to the specified domain

plt.show()
```

The plot() method connects all points as specified by the domain and the variable relationship, therefore the discontinuity is not obvious. To emphasize such discontinuity, we could just split the domain in two, as explained next.

```python
x1 = np.linspace(-5,-1/5,100) # Specify independent variable as a domain 1 with interval step size
x2 = np.linspace(1/5,5,100) # Specify independent variable as a domain 2 with interval step size

y1 = 1/x1 # Define relationship between independent and dependent variable
y2 = 1/x2 # Define relationship between independent and dependent variable

plt.plot(x1, y1) # The graph is restricted to the specified domain
plt.plot(x2, y2) # The graph is restricted to the specified domain

plt.show()
```

```python
x1 = np.linspace(-5,-1/5,100) # Specify independent variable as a domain 1 with interval step size
x2 = np.linspace(1/5,5,100) # Specify independent variable as a domain 2 with interval step size

y1 = 1/x1 # Define relationship between independent and dependent variable
y2 = 1/x2 # Define relationship between independent and dependent variable

l = np.linspace(-5,5,2) # Line
o = np.linspace(0,0,2) # Line

plt.plot(x1, y1) # The graph is restricted to the specified domain
plt.plot(x2, y2) # The graph is restricted to the specified domain

plt.plot(l,o) # Horizontal Asymptote
plt.plot(o,l) # Vertical Asymptote

plt.show()
```

$$ y=\frac{1}{x} $$

```python
x1 = np.linspace(-5,-1/5,100) # Specify independent variable as a domain 1 with interval step size
x2 = np.linspace(1/5,5,100) # Specify independent variable as a domain 2 with interval step size

y1 = 1/x1 # Define relationship between independent and dependent variable
y2 = 1/x2 # Define relationship between independent and dependent variable

l = np.linspace(-5,5,2) # Line
o = np.linspace(0,0,2) # Line

plt.plot(x1, y1,'b-') # The graph is restricted to the specified domain
plt.plot(x2, y2,'b-') # The graph is restricted to the specified domain

plt.plot(l,o,'k--') # Horizontal Asymptote
plt.plot(o,l,'k--') # Vertical Asymptote

plt.show()
```

$$ y = \frac{k}{x-p}+q $$

```python
# Parameters

p = 2
q = 3
k = 1

# Vertical and Horizontal Asymptotes

VA = p
HA = q

x1 = np.linspace(-5+VA,-1/5+VA,100) # Translate by VA units
x2 = np.linspace(1/5+VA,5+VA,100) # Translate by VA units

y1 = k/(x1-VA) + HA # Define relationship between independent and dependent variable
y2 = k/(x2-VA) + HA # Define relationship between independent and dependent variable

l = np.linspace(-5+VA,5+VA,2) # Translate by VA units
o = np.linspace(0+HA,0+HA,2) # Translate by HA units

r = np.linspace(0+VA,0+VA,2) # Translate by VA units
s = np.linspace(-5+HA,5+HA,2) # Translate by HA units

plt.plot(x1, y1,'b-') # The graph is restricted to the specified domain
plt.plot(x2, y2,'b-') # The graph is restricted to the specified domain

plt.plot(l,o,'k--') # Horizontal Asymptote

plt.plot(r,s,'k--') # Vertical Asymptote

plt.show()
```
